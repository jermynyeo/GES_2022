```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('data\GES.csv')
df.head()
```

```python
df.dtypes
```

```python
df['basic_monthly_mean'] = df['basic_monthly_mean'].astype(int)
```

```python
df = df.drop(df[df['basic_monthly_mean'] == 'na'].index)
```

```python
fig = px.scatter(df, x="basic_monthly_mean", y="employment_rate_overall", color="university", hover_name="degree", log_x=True, size_max=60)
fig.show()
```
